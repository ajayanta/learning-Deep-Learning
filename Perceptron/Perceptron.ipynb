```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import Perceptron

# Load dataset
df = pd.read_csv(r"C:\Users\KIIT\Desktop\100days deep learning\perceptron\placement .csv")

# Visualize the data
sns.scatterplot(x=df['cgpa'], y=df['resume_score'], hue=df['placed'], palette='Set1')

# Prepare data
x = df.iloc[:, 0:2]
y = df.iloc[:, -1]

# Train Perceptron
p = Perceptron()
p.fit(x, y)

# Get weights and bias
w1, w2 = p.coef_[0]
b = p.intercept_[0]

# Plot decision boundary: w1*x + w2*y + b = 0 => y = -(w1*x + b)/w2
x_vals = np.linspace(df['cgpa'].min(), df['cgpa'].max(), 100)
y_vals = -(w1 * x_vals + b) / w2

# Plot the line
plt.plot(x_vals, y_vals, color='green', label='Decision Boundary')

# Labels and grid
plt.xlabel('CGPA')
plt.ylabel('Resume Score')
plt.title('Perceptron Decision Boundary')
plt.legend()
plt.grid(True)
plt.show()
```
